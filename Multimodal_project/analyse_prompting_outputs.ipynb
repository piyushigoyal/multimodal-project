## HTML output

```python
from IPython.display import HTML
from PIL import Image
import base64
from io import BytesIO
import jsonlines
import os

# === Helper functions ===
def load_jsonl(path):
    with jsonlines.open(path) as reader:
        return list(reader)

def pil_to_base64_html(img, width=300):
    buffer = BytesIO()
    img.save(buffer, format="PNG")
    base64_img = base64.b64encode(buffer.getvalue()).decode("utf-8")
    return f'<img src="data:image/png;base64,{base64_img}" width="{width}"/>'

def get_filename(path):
    return os.path.basename(path)

# === Load all files ===
base_dir = "/Users/user/Library/CloudStorage/OneDrive-King'sCollegeLondon/PycharmProjects/multimodal-project/Multimodal_project/"
master_path = os.path.join(base_dir, "mathvista_data/testmini/master_output.jsonl")
interactive_path = os.path.join(base_dir, "mathvista_data/testmini/inter_cot_outputs.jsonl")
prompt_path = os.path.join(base_dir, "mathvista_data/testmini/prompt_outputs.jsonl")

master_data = load_jsonl(master_path)
interactive_data = load_jsonl(interactive_path)
prompt_data = load_jsonl(prompt_path)

interactive_filename = get_filename(interactive_path)
prompt_filename = get_filename(prompt_path)
master_filename = get_filename(master_path)

# === Build HTML blocks ===
html_blocks = []
for entry in master_data[:100]:  # Adjust number of examples if needed
    pid = entry['pid']
    img_path = os.path.join(base_dir, entry['image_path'])
    img_html = pil_to_base64_html(Image.open(img_path))

    # Match entries
    interactive_entry = next((item for item in interactive_data if item["pid"] == pid), None)
    prompt_entry = next((item for item in prompt_data if item["pid"] == pid), None)

    # === Compose HTML ===
    html = f"""
    <div style="margin-bottom: 50px; border-bottom: 1px solid #ccc; padding-bottom: 20px;">
        <h2>Query (PID: {pid})</h2>
        <pre style="white-space: pre-wrap;">{entry['query']}</pre>
        {img_html}
        <p><strong>Ground Truth:</strong> {entry['ground_truth_answer']}</p>

        <h3>From {master_filename}</h3>
        <p><strong>Generated Answer:</strong><br>{entry['generated_answer']}</p>
        <p><strong>Extracted Answer:</strong> {entry['extracted_answer']}</p>
    """

    if interactive_entry:
        html += f"""
        <h3>From {interactive_filename}</h3>
        <p><strong>Generated Answer:</strong><br>{interactive_entry['generated_answer']}</p>
        <p><strong>Extracted Answer:</strong> {prompt_entry['extracted_answer']}</p>
        """
        if 'revised_generated_answer' in interactive_entry:
            html += f"<p><strong>Revised Generated Answer:</strong><br>{interactive_entry['revised_generated_answer']}</p>"

        if 'revised_extracted_answer' in interactive_entry:
            html += f"<p><strong>Revised Extracted Answer:</strong> {interactive_entry['revised_extracted_answer']}</p>"
    else:
        html += f"<p><em>No entry found in {interactive_filename} for pid={pid}</em></p>"

    if prompt_entry:
        html += f"""
        <h3>From {prompt_filename}</h3>
        <p><strong>Generated Answer:</strong><br>{prompt_entry['generated_answer']}</p>
        <p><strong>Extracted Answer:</strong> {prompt_entry['extracted_answer']}</p>
        """
    else:
        html += f"<p><em>No entry found in {prompt_filename} for pid={pid}</em></p>"

    html += "</div>"
    html_blocks.append(html)

# === Save as full HTML ===
full_html = "<html><body style='font-family: sans-serif;'>\n" + "\n".join(html_blocks) + "\n</body></html>"

output_path = os.path.join(base_dir, "sample_report.html")
with open(output_path, "w", encoding="utf-8") as f:
    f.write(full_html)

print(f"Report saved to: {output_path}")
```
